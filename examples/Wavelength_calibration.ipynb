# Echelle imaging and getting spectra

# Wavelength Calibration

```python
from echelle_pkg.calibration import WavelengthCalibration
```

```python
files_cmos = {
    "orders": "pattern.txt",    # your pattern.txt
    "sphr": "20250123_sphere_100ms.tif",    # your sphere image (積分球)
    "bkgr": "20250123_sphereback_100ms.tif",    # your background image of sphere
}

path = 'calibration_file_examples'   # your calibration folder name
```

```python
wcb = WavelengthCalibration(path, files_cmos, crop=[100,1850], crop2=[20,1095])  # your crop range (切り取る画像ピクセル範囲を設定)
wcb.setup_calibration()
wcb.setup_image('20250123_sphere_100ms.tif')    # your sphere image
```

```python
# 画像確認
wcb.plot_cut_image()
wcb.plot_frame()
```

```python
# オーダースペクトル計算
order_spectra = wcb.calculate_order_spectra()
```

```python
# 波長校正
wfits, cb_xlist, cb_ylist = wcb.fit_wavelength('wavelength_hayashi.txt')    # your pixel-wavelength convert text
```
